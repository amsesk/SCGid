```python
import plotly.express as px
import numpy as np
from scgid.scripts.infotable import InfoTable
it = InfoTable()
it.load("../test_data/devtest_scgid_output/gct/devtest.infotable.tsv")
it.df = it.df.assign(scaled_evalue = np.log(it.df.evalue))

fig = px.scatter(x=np.log(it.df.coverage),
                 y=it.df.gc,
                 color=it.df.parse_lineage,
                 size=abs(it.df.scaled_evalue),
                 opacity=0.4
                )
fig.update_xaxes(title="log(Coverage)")
fig.update_yaxes(title="GC Content")
fig.show()
```

```python
fig = px.scatter(x=np.log(it.df.coverage),
                 y=it.df.gc,
                 color=it.df.pertinent_taxlvl,
                 size=abs(it.df.scaled_evalue),
                 opacity=0.4
                )
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

t_gc_counts,t_gc_div = np.histogram(it.df[it.df.parse_lineage == "target"].gc, bins=1000)
t_cov_counts,t_cov_div = np.histogram(it.df[it.df.parse_lineage == "target"].coverage, bins=1000)

nt_gc_counts,nt_gc_div = np.histogram(it.df[it.df.parse_lineage == "nontarget"].gc, bins=1000)
nt_cov_counts,nt_cov_div = np.histogram(it.df[it.df.parse_lineage == "nontarget"].coverage, bins=1000)

fig = make_subplots(rows=2, cols=2, column_widths=[0.8,0.2], row_heights=[0.2,0.8], 
                    shared_xaxes=True, shared_yaxes=True, vertical_spacing=0.01, horizontal_spacing=0.01)  
fig.add_trace ( go.Scatter(x=np.log(it.df[it.df.parse_lineage == "target"].coverage),
                y=it.df[it.df.parse_lineage == "target"].gc, mode="markers", marker=dict(color="blue"),
                          opacity=0.5), row=2, col=1)
fig.add_trace ( go.Scatter(x=np.log(it.df[it.df.parse_lineage == "nontarget"].coverage),
                y=it.df[it.df.parse_lineage == "nontarget"].gc, mode="markers", marker=dict(color="red"),
                          opacity=0.5), row=2, col=1)

fig.add_trace( go.Scatter(x=np.log(t_cov_div), y=t_cov_counts, line=dict(color="blue")), row=1, col=1)
fig.add_trace( go.Scatter(x=t_gc_counts, y=t_gc_div, line=dict(color="blue")), row=2, col=2)

fig.add_trace( go.Scatter(x=np.log(nt_cov_div), y=nt_cov_counts, line=dict(color="red")), row=1, col=1)
fig.add_trace( go.Scatter(x=nt_gc_counts, y=nt_gc_div, line=dict(color="red")), row=2, col=2)

fig.show()
```
